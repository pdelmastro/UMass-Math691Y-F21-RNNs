# Setup + Function Definitions

```python
# Imports
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

# Local
from util import my_odeint
```

```python
"""
Fully connected feed-forward neural network 
"""

class FCN(nn.Module):
    
    def __init__(self, N_INPUT, N_OUTPUT, N_HIDDEN, N_LAYERS, activation='tanh'):
        super().__init__()
        # Select the activation function
        if activation == 'tanh':
            activation = nn.Tanh
        elif activation == 'relu':
            activation = nn.ReLU
        elif activation == 'sine':
            activation = torch.sin
        self.fcs = nn.Sequential(*[
                        nn.Linear(N_INPUT, N_HIDDEN),
                        activation()])
        self.fch = nn.Sequential(*[
                        nn.Sequential(*[
                            nn.Linear(N_HIDDEN, N_HIDDEN),
                            activation()]) for _ in range(N_LAYERS-1)])
        self.fce = nn.Linear(N_HIDDEN, N_OUTPUT)
        
    def forward(self, x):
        x = self.fcs(x)
        x = self.fch(x)
        x = self.fce(x)
        return x
```

```python
"""
Functions for computing derivatives of the network's output wrt its input
"""

def compute_dx_dt(t, x):
    """
    Computes dx/dt for the given inputs 't' and outputs 'x'
    """
    return torch.autograd.grad(x, t, torch.ones_like(x), create_graph=True)[0]


def compute_d2x_dt2(t, dx_dt):
    """
    Computes d^2x/dt^2 for the inputs 't' and outputs 'x' (not a parameter here)
    Must be given dx_dt, the derivative dx/dt computed from compute_dx_dt()
    """
    return torch.autograd.grad(dx_dt, t, torch.ones_like(dx_dt), create_graph=True)[0]
```

```python
"""
Function to train neural networks with or without the PINN loss function
"""

def train(network, 
          t_data=None, x_data=None, dx_dt_data=None,
          t_physics=None, f_physics=None,
          n_iters=1000, lr=1e-3, alpha=1e-4, 
          beta=1, verbose=None,
          time_scaling=1
    ):
    """
    ARGUMENTS
    
        network      :  FCN network to be trained
        
        t_data       :  (t, x)-pairs that the network will be trained on
        x_data  
        
        t_physics    :  Times at which the network's "physics" will be evaluated
        
        n_iters      :  Number of training iterations
        
        f_phys       :  Function that computes the physical quantities that 
                        should be minimized
        
        lr           :  Learning rate
        
        alpha        :  Controls the influence of the physics loss.
                        Overall loss is computed as 
                            loss = data_loss + alpha * physics_loss
                            
        beta         :  Controls the influence of the dx/dt loss
        
        tau          :  Input times are rescaled to t/tau 
                        when computing ???
    
    RETURNS
    """
    # Create the optimizer
    optimizer = torch.optim.Adam(network.parameters(),lr=1e-3)
    
    # Arrays to store the losses
    data_losses = torch.zeros(n_iters)
    physics_losses = torch.zeros(n_iters)

    # Training loop
    for i in range(n_iters):
        # Clear the parameter gradients
        optimizer.zero_grad()
        
        # Variable to store the network's total loss
        loss = 0
        
        # Compute the network's loss on data
        if t_data is not None and x_data is not None:
            # x-loss
            xh_data = network(t_data)
            data_loss = torch.mean((xh_data-x_data)**2)
            # dx/dt-loss
            if dx_dt_data is not None:
                dxh_dt = compute_dx_dt(t_data, xh_data)
                data_loss += beta * torch.mean((dxh_dt-dx_dt_data)**2)
            # Update the loss and save the data loss
            loss += data_loss
            data_losses[i] = data_loss.detach()
                
        
        # Compute the network's "physics" loss
        if t_physics is not None and f_physics is not None:
            physics = f_physics(t_physics, network(t_physics))
            physics_loss = torch.mean(physics ** 2)
            loss += alpha * physics_loss
            # Save the physics loss
            physics_losses[i] = physics_loss.detach()
            
        
        # Update the network's parameter
        loss.backward()
        optimizer.step()
        
        if verbose and ((i+1) % verbose) == 0:
            print('Epoch %d/%d, Loss=%.9f' % (i+1,n_iters,loss))
            
    # Return the loss arrays as np array
    return data_losses.numpy(), physics_losses.numpy()
```

```python
"""
Function to plot the results
"""
def plot_result(x,y,x_data,y_data,yh,xp=None):
    "Pretty plot training results"
    plt.figure(figsize=(8,4))
    plt.plot(x,y, color="grey", linewidth=2, alpha=0.8, label="Exact solution")
    plt.plot(x,yh, color="tab:blue", linewidth=4, alpha=0.8, label="Neural network prediction")
    plt.scatter(x_data, y_data, s=60, color="tab:orange", alpha=0.4, label='Training data')
    if xp is not None:
        plt.scatter(xp, -0*torch.ones_like(xp), s=60, color="tab:green", alpha=0.4, 
                    label='Physics loss training locations')
#     if T_train is not None:
#         plt.axvline(x=T_train)
    l = plt.legend(loc=(1.01,0.34), frameon=False, fontsize="large")
    plt.setp(l.get_texts(), color="k")
    # plt.xlim(-0.05, 1.05)
    # plt.ylim(-1.1, 1.1)
    plt.axis("off")
```

```python
"""
Function to compute the activations of all neurons in the network
"""

def compute_activations(network, times):
    activation_by_layer = []
    
    layers = [network.fcs] + [l for l in network.fch] + [network.fce]
    
    with torch.no_grad():
        for i, layer in enumerate(layers):
            if i == 0:
                inp = times
            else:
                inp = activation_by_layer[-1]
            if i+1 != len(layers):
                out = layer[0](inp)
                out = layer[1](out)
            else:
                out = layer(out)
            activation_by_layer.append(out)
    
    return activation_by_layer
```

# Damped Harmonic Oscillator - Forward Problem

```python
"""
Physics function for the harmonic oscillator
"""
def sho_physics(t, x, mu, k):
    # Compute dx/dy and d^2x/dt^2
    dx_dt = compute_dx_dt(t, x)
    d2x_dt2 = compute_d2x_dt2(t, dx_dt)
    # Compute the SHO physics
    return d2x_dt2 + mu*dx_dt + k*x

"""
SHO vector field
"""
def F_sho(x, args):
    mu,k = args
    return np.array([
        x[1],
        -k * x[0] - mu * x[1]
    ])
```

```python
"""
Example training the network and visualizing it's predictions
"""

# Parameters
mu, k = 4, 400
x0 = 1
v0 = 1
t_physics = torch.linspace(0,1,30).view(-1,1).requires_grad_(True)
lr = 1e-3
n_iters = 50000
alpha = 1e-3
beta = 1
verbose = int(n_iters / 25) 
activation = 'tanh'

# Random seed
"""
691658 - works for 3x8 tanh neurons
"""
seed = np.random.randint(1e6)
print('Random seed:', seed)
torch.manual_seed(seed)

# Package x0 and v0 up into data
t_data = torch.zeros(1).view(-1,1).requires_grad_(True)
x_data = torch.ones(1).view(-1,1) * x0
dx_dt_data = torch.ones(1).view(-1,1) * v0

# Define the physics function
f_physics = lambda t,x: sho_physics(t,x,mu,k)

# Create the network
network = FCN(1,1,8,3, activation=activation)

# Run the Training loop
train(network, 
    t_data, x_data, dx_dt_data,
    t_physics, f_physics,
    n_iters, lr, 
    alpha, beta,
    verbose
)
```

```python
# Exact solution parameters
dt = 1e-3
T = 1.5
s0 = np.array([x0,v0])

# Generate the exact solution
x_exact = my_odeint(F_sho, dt, T, s0, args=(mu,k))

# Generate the network's solution
t_viz = dt * torch.arange(x_exact.shape[0]).view(-1,1)
x_net = network(t_viz).detach()

# Visualize
plot_result(
    t_viz, x_exact[:,0], 
    t_data.detach(), x_data.detach(), 
    x_net,
    t_physics.detach()
)
plt.ylim(-1.1, 1.1)
plt.axvline(x=1, color='black', linewidth=2, linestyle='--',zorder=-10)
plt.show()
```

# Damped Harmonic Oscillator - Inverse Problem

```python
"""
Parameters
"""
# ODE Parameters
mu, k = 4, 400
x0 = 1
v0 = 1
dt = 1e-3
T = 1
T_viz = 1.5
n_samples = 10

# Training parameters
n_t_physics = 30
lr = 1e-3
n_iters = 20000
alpha = 1e-4
beta = 0
verbose = int(n_iters / 25) 

"""
Setup and Training
"""

# Random seed
seed = np.random.randint(1e6)
print('Random seed:', seed)
torch.manual_seed(seed)

# Generate the exact solution
x_exact = torch.from_numpy(
    my_odeint(F_sho, dt, T_viz, s0, args=(mu,k))
)
times = dt * torch.arange(x_exact.shape[0]).view(-1,1)

# Sample x_exact to generate the training data
max_ind = int(T/dt)
sample_inds = torch.from_numpy(
    np.random.choice(
        np.arange(max_ind), size=n_samples, replace=False
    )
)
t_data = times[sample_inds].view(-1,1)
x_data = x_exact[sample_inds,0].view(-1,1)

# Define the physics function
f_physics = lambda t,x: sho_physics(t,x,mu,k)

# Times when the network's physics will be eval'd
t_physics = torch.linspace(0,T,n_t_physics).view(-1,1).requires_grad_(True)

# Create the network
network = FCN(1,1,32,3)

# Run the Training loop
train(network, 
    t_data, x_data, None, # no velocity data
    t_physics, f_physics,
    n_iters, lr, 
    alpha, beta,
    verbose
)
```

```python
# Generate the network's solution at all times
x_net = network(times).detach()

# Visualize
plot_result(
    times, x_exact[:,0], 
    t_data.detach(), x_data.detach(), 
    x_net,
    t_physics.detach()
)
plt.ylim(-1,1)
plt.show()
```

# Damped Nonlinear Oscillator - Forward Problem

```python
"""
Physics function for the nonlinear oscillator
"""
def nlo_physics(t, x, b, tau=1):
    # Compute dx/dy and d^2x/dt^2
    dx_dt = compute_dx_dt(t, x)
    d2x_dt2 = compute_d2x_dt2(t, dx_dt)
    # Compute the SHO physics
    return d2x_dt2 + tau * b*dx_dt + (tau ** 2) * torch.sin(x)

"""
Nonlinear oscillator vector field
"""
def F_nlo(x, args):
    b = args[0]
    return np.array([
        x[1],
        -1 * np.sin(x[0]) - b * x[1]
    ])
```

```python
# Parameters
b = 0.1
tau = 20
x0 = 1
v0 = 1
n_t_physics = 30

lr = 1e-4
n_iters = 30000
alpha = 1e-3
beta = 1/tau
verbose = int(n_iters / 25) 

# Random seed
"""
Successes
1. 67935 with b = 0.1, tau = 20, alpha = 1e-4, beta = 1/tau, n_iters=20k, n_physics=30, x0=v0=1
2. 109197 same as 1. but lr=5e-3
3. 813383 same as 1. but lr=3e-3 and n_iters=30k 
4. 631100 lr = 1e-4, n_iters = 30000, alpha = 1e-5, not successful but big jump in loss

Notes
- Needed to squash to from [0, 20] to [0, 1] by change of variable using tau
- Needed to pick the right ratio of alpha, beta, and lr 
"""
seed = np.random.randint(1e6)
print('Random seed:', seed)
torch.manual_seed(seed)

# Package x0 and v0 up into data
t_data = torch.zeros(1).view(-1,1).requires_grad_(True)
x_data = torch.ones(1).view(-1,1) * x0
dx_dt_data = torch.ones(1).view(-1,1) * v0 * tau

# Define the physics function
f_physics = lambda t,x: nlo_physics(t,x,b,tau)
t_physics = torch.linspace(0,1,n_t_physics).view(-1,1).requires_grad_(True)

# Create the network
nlo_network = FCN(1,1,32,3)

# Run the Training loop
data_losses, physics_losses = train(nlo_network, 
    t_data, x_data, dx_dt_data,
    t_physics, f_physics,
    n_iters, lr, 
    alpha, beta,
    verbose
)
```

```python
# Learning curve
plt.figure(figsize=(12,8))
plt.title('Learning curves', fontsize=16)
plt.xlabel('Training Time (# parameter updates)', fontsize=16)
# plt.ylabel('Log(Loss)', fontsize=16)
plt.plot(np.log(data_losses[:15000]), '-b', label='log(data loss)')
plt.plot(np.log(alpha * physics_losses[:15000]), '-r', label=r'log($\alpha\cdot$ physics loss)')
plt.legend(fontsize=16)
plt.show()
```

```python
# Exact solution parameters
dt = 1e-3
tau = 20
T_viz = 30
s0 = np.array([x0,v0])

# Generate the exact solution
x_exact = my_odeint(F_nlo, dt, T_viz, s0, args=(b,))

# Generate the network's solution
t_viz = dt * torch.arange(x_exact.shape[0]).view(-1,1)
with torch.no_grad():
    x_net = nlo_network(t_viz/tau).detach()

# Visualize
plot_result(
    t_viz, x_exact[:,0], 
    t_data.detach(), x_data.detach(), 
    x_net, tau * t_physics.detach()
)
plt.axvline(x=tau, color='black', linewidth=2, linestyle='--',zorder=-10)
plt.text(
    tau,1,'t=T',fontsize=16, ha='center',
    bbox = dict(facecolor = 'wheat', alpha = 0.9)
)
plt.show()
```

# Vander-Pol Oscillator - Forward Problem

```python
"""
Physics function for the vanderpol oscillator
"""
def vander_pol_physics(t, x, mu, tau):
    # Compute dx/dy and d^2x/dt^2
    dx_dt = compute_dx_dt(t, x)
    d2x_dt2 = compute_d2x_dt2(t, dx_dt)
    # Compute the Vander Pol osc physics
    return d2x_dt2 + tau * mu*(x*x - 1)*dx_dt + (tau ** 2) * x

"""
Nonlinear oscillator vector field
"""
def F_vander_pol(x, args):
    mu = args[0]
    return np.array([
        x[1],
        -mu * (x[0] * x[0] - 1) * x[1] - x[0]
    ])
```

```python
# Parameters
mu = 1
x0 = 1
v0 = 1
tau = 30

lr = 3e-3
n_iters = 20000
alpha = 1e-4
beta = 1 / tau
verbose = int(n_iters / 10) 
n_t_physics = 30

# Random seed
seed = np.random.randint(1e6) # 379741
print('Random seed:', seed)
torch.manual_seed(seed)

# Package x0 and v0 up into data
t_data = torch.zeros(1).view(-1,1).requires_grad_(True)
x_data = torch.ones(1).view(-1,1) * x0
dx_dt_data = torch.ones(1).view(-1,1) * v0 * tau

# Define the physics function
f_physics = lambda t,x: vander_pol_physics(t,x,b,tau)

# Create the network
vdp_network = FCN(1,1,32,3)
# Start from a copy of the nonlinear oscillator
# that learned the undamped oscillations
# vdp_network = copy.deepcopy(nlo_network_copy)

# Scale the weights of the network
g = 1.0
layers = [vdp_network.fcs[0], vdp_network.fce] + [l[0] for l in vdp_network.fch]
for layer in layers:
    layer.weight.data = layer.weight * g

    
# Basic training
# t_physics = torch.linspace(0,1,n_t_physics).view(-1,1).requires_grad_(True)
# data_losses, physics_losses = train(vdp_network, 
#     t_data, x_data, dx_dt_data,
#     t_physics, f_physics,
#     n_iters, lr, 
#     alpha, beta,
#     verbose
# )

# Curriculum learning approach
Ts_physics = [0.05, 0.1, 0.15, 0.2, 0.4, 0.6, 1]
for k, T_physics in enumerate(Ts_physics):
    print('Curriculum {}/{} T_physics={}'.format(
        k+1, len(Ts_physics), T_physics)
    )
    t_physics = torch.linspace(0,T_physics,n_t_physics).view(-1,1).requires_grad_(True)
    train(vdp_network, 
        t_data, x_data, dx_dt_data,
        t_physics, f_physics,
        n_iters, lr, 
        alpha, beta,
        verbose
    )
```

```python
# Exact solution parameters
dt = 1e-3
T_viz = tau
s0 = np.array([x0,v0])

# Generate the exact solution
x_exact = my_odeint(F_vander_pol, dt, T_viz, s0, args=(mu,))

# Generate the network's solution
t_viz = dt * torch.arange(x_exact.shape[0]).view(-1,1)
with torch.no_grad():
    x_net = vdp_network(t_viz/tau).detach()

# Visualize
plot_result(
    t_viz, x_exact[:,0], 
    t_data.detach(), x_data.detach(), 
    x_net, tau * t_physics.detach()
)
plt.show()
```
