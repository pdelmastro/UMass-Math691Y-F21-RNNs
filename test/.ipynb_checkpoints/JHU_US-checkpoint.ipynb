Testing the function ```JHU_US()``` in ```src/load_data.py```

````python
# Change root directory
import sys, os
sys.path.append(os.path.abspath('../'))

# Import the JHU_US function
from src.load_data import JHU_US

# Useful imports
import numpy as np
import matplotlib.pyplot as plt

import matplotlib as mlp
mlp.rc('font', size=16)

# Location of the data directory
datadir = '../datasets/jhu'
````

# MA data

````python
u_MA = JHU_US(
    dtype='cases_daily', 
    seq_len=None, 
    granularity='state',
    select_state='MA',
    rescaling='population', # 'None',
    squash_to=None,
    smooth=7,# None, 
    datadir=datadir,
    lib='np'
)

plt.figure(figsize=(12,6))
plt.title('MA cases daily')
plt.xlabel('Time t (# days since start of pandemic)')
plt.ylabel('Daily # Cases')
plt.plot(u_MA.reshape(-1))
plt.show()
````

````python
counties, u_MA_counties = JHU_US(
    dtype='cases_daily', 
    seq_len=None, 
    granularity='county',
    select_state='MA',
    rescaling='population',
    squash_to=None,
    smooth=7,# None, 
    datadir=datadir,
    lib='np'
)

plt.figure(figsize=(16,8))
plt.title('MA daily cases by county (smoothed, rescaled)')
plt.xlabel('Time t (# days since start of pandemic)')
plt.ylabel('# Daily Cases')
for i in range(counties.shape[0]):
    plt.plot(u_MA_counties[i], label=counties[i])
plt.legend()
plt.show()
````
